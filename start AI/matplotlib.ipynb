```python
a = np.array([1, 3, 5, 6, 8, 9, 10, 11])
b = np.array([3, 4, 5, 7, 8, 9, 11, 10])
```

```python
plt.plot(a, b, marker = "s", ms = 4, markeredgecolor = "green", color = "red", linewidth = 3)
plt.grid()
plt.xlabel("Week days")
plt.ylabel("Profit in lakh")
plt.title("My Mothly revnew")
```

```python
# Scatter : 

a = np.array([1, 3, 5, 6, 8, 9, 10, 11])
b = np.array([3, 4, 5, 7, 8, 9, 11, 10])
```
