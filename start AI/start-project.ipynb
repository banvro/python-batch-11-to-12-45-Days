```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("heart.csv")
```

```python
data
```

```python
data.info()
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lbe = LabelEncoder()
```

```python
zx = ["hlo", "apple", "banana", "orange", "banana", "banana", "hlo"]
```

```python
er = lbe.fit_transform(zx)
```

```python
er
```

```python
data["Sex"] = lbe.fit_transform(data["Sex"])
```

```python
data
```

```python
data["ChestPainType"] = lbe.fit_transform(data["ChestPainType"])
data["RestingECG"] = lbe.fit_transform(data["RestingECG"])
data["ExerciseAngina"] = lbe.fit_transform(data["ExerciseAngina"])
data["ST_Slope"] = lbe.fit_transform(data["ST_Slope"])
```

```python
data
```

```python
data.head()
```

```python
x = data.drop(columns="HeartDisease")
y = data["HeartDisease"]
```

```python
x
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 20, test_size = 0.2)
```

```python
x_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
mdl = LogisticRegression()
```

```python
mdl.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
x_test
```

```python
y_pred = mdl.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
accuracy_score(y_pred, y_test)
```
