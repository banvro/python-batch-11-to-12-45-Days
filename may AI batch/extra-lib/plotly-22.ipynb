```python
import plotly.graph_objects as go
```

```python
x = [12, 23, 45, 67, 34, 23, 45, 65]
y = [12, 34, 45, 67, 78, 80, 84, 100
```

```python
fig = go.Figure(data = go.Scatter(x = x, y = y))
```

```python
fig = go.Figure(data = go.Scatter(x = x, y = y))
fig.update_layout(title = "My Data", xaxis_title = "this is is x axis data", yaxis_title = "this is y asix data")
```

```python
fig = go.Figure(
    data = go.Scatter(x = x, y = y, mode = "markers"),
)

fig.update_layout(
    title = "My Data", 
    xaxis_title = "this is is x axis data", 
    yaxis_title = "this is y asix data"
)
```

```python
fig = go.Figure(
    data = go.Scatter(x = x, y = y, mode = "lines"),
)

fig.update_layout(
    title = "My Data", 
    xaxis_title = "this is is x axis data", 
    yaxis_title = "this is y asix data"
)
```

```python
fig = go.Figure(
    data = go.Scatter(x = x, y = y, mode = "markers+lines"),
)

fig.update_layout(
    title = "My Data", 
    xaxis_title = "this is is x axis data", 
    yaxis_title = "this is y asix data"
)
```

```python
fig = go.Figure(
    data = go.Scatter(x = x, y = y, mode = "markers+lines", marker = dict(color = "red", size = 14)),
)
fig.show()
```

```python
x = [12, 23, 45, 67, 34, 23, 45, 65]
y = [12, 34, 45, 67, 78, 80, 84, 100
```

```python
fig = go.Figure(
    data = go.Scatter(x = x, y = y, 
                      mode = "markers+lines", 
                      marker = dict(color = "blue", size = 14),
                      hovertext = "this is text", 
                      hoverinfo = "text + x + y",
                      line = dict(color = "green", width = 3, dash = "dot")
                     ),
)
# ['solid', 'dot', 'dash', 'longdash', 'dashdot', 'longdashdot']


# for i in range()


fig.add_annotation(x = x[5], y = y[5], text = "hello", showarrow = False)
fig.show()
```

```python
fig = go.Figure(
    data = go.Scatter(x = x, y = y, 
                      mode = "markers+lines", 
                      marker = dict(color = "blue", size = 14),
                      hovertext = "this is text", 
                      hoverinfo = "text + x + y",
                      line = dict(color = "green", width = 3, dash = "dot")
                     ),
)
# ['solid', 'dot', 'dash', 'longdash', 'dashdot', 'longdashdot']


# for i in range()


fig.add_annotation(x = x[5], y = y[5], text = "hello", showarrow = True, arrowcolor = "red")
fig.show()
```

```python
fig = go.Figure(
    data = go.Scatter(x = x, y = y, 
                      mode = "markers+lines", 
                      marker = dict(color = "blue", size = 14),
                      hovertext = "this is text", 
                      hoverinfo = "text + x + y",
                      line = dict(color = "green", width = 3, dash = "dot")
                     ),
)

for i in range(len(x)):
    fig.add_annotation(x = x[i], y = y[i], text = "hello")

fig.show()
```
