```python
df = pd.DataFrame({
    "Age" : [12, 34, 23, 45, 23, 34, 23, 45, 23], 
    "fvrt_color" : ["blue", "red", "green", "pink", "green", "blue", "pink", "blue", "blue"],
    "number" : [89263482364, 98263498, 9823648, 982634892, 983648923, 98364982, 8725384234, 892364923, 98236489]
})
```

```python
df
```

```python
oe = OrdinalEncoder(categories = [["red", "green", "pink", "blue"]])
```

```python
ordinalecn = oe.fit_transform(df[["fvrt_color"]])
```

```python
df["color_o_e"] = ordinalecn
```

```python
df = pd.DataFrame({
    "Age" : [12, 34, 23, 45, 23, 34, 23, 45, 23], 
    "fvrt_color" : ["blue", "red", "green", "pink", "green", "blue", "pink", "blue", "blue"],
    "number" : [89263482364, 98263498, 9823648, 982634892, 983648923, 98364982, 8725384234, 892364923, 98236489]
})
```
