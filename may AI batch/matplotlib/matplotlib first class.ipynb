```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# plot
```

```python
xyz = np.array([12, 34, 56, 60, 70])
abc = np.array([100, 300, 400, 350, 360])
```

```python
plt.plot(xyz, abc)
```

```python
# ms => marker mize

plt.plot(xyz, abc, marker = "*", ms = 12)
```

```python
# change marker color

# mec -> marcker edge color
# mfc -> marker face color

plt.plot(xyz, abc, marker = "*", ms = 15, mec = "red", mfc = "yellow")
```

```python
plt.plot(xyz, abc, marker = "*", ms = 15, mec = "red", mfc = "yellow")
plt.xlabel("User Age..") #for horizatal
plt.ylabel("Income in Thousends")  #vertical
plt.title("User Income Info", color = "red", loc = "center")

# loc option:
# left
# right
# center
```

```python
# use grid in background
```

```python
plt.plot(xyz, abc, marker = "*", ms = 15, mec = "red", mfc = "yellow")
plt.xlabel("User Age..")
plt.ylabel("Income in Thousends")
plt.title("User Income Info", color = "red")
plt.grid()
```

```python
age = np.array([12, 20, 45, 50, 70, 89])
income = np.array([10, 30, 45, 56, 34, 60])
```

```python
plt.plot(age, income, marker = ".", ms = 20, mfc = "red", mec = "green")
plt.xlabel("User Age")
plt.ylabel("User Income in k")
plt.title("User incorom info")
plt.grid()
```

```python
plt.plot(age, income, marker = ".", ms = 20, mfc = "yellow", mec = "green", color = "red", linewidth = 3)
plt.xlabel("User Age")
plt.ylabel("User Income in k")
plt.title("User incorom info")
plt.grid()
```

```python
plt.plot(age, income, marker = ".", ms = 20, mfc = "yellow", mec = "green", color = "red", linewidth = 3, linestyle = "dashed")
plt.xlabel("User Age")
plt.ylabel("User Income in k")
plt.title("User incorom info")
plt.grid()
```

```python
plt.plot(age, income, marker = ".", ms = 20, mfc = "yellow", mec = "green", color = "red", linewidth = 3, linestyle = "dotted")
plt.xlabel("User Age")
plt.ylabel("User Income in k")
plt.title("User incorom info")
plt.grid()
```

```python
plt.plot(age, income, marker = ".", ms = 20, mfc = "yellow", mec = "green", color = "red", linewidth = 3, linestyle = "dashdot")
plt.xlabel("User Age")
plt.ylabel("User Income in k")
plt.title("User incorom info")
plt.grid()
```

```python
plt.plot(age, income, marker = ".", ms = 20, mfc = "yellow", mec = "green", color = "red", linewidth = 3, linestyle = "dotted")
plt.xlabel("User Age")
plt.ylabel("User Income in k")
plt.title("User incorom info")
plt.grid(color = "green", linestyle = "dotted", linewidth = 2)
```
