```python
product_name = ["Hanky", "Shoose", "Creems", "Shirts", "Jeens"]
record_2023 = [20, 40, 30, 50, 40]
record_2024 = [30, 50, 40, 60, 80]
```

```python
plt.bar(product_name, record_2023, color = "red", label = "record 2023")
plt.bar(product_name, record_2024, color = "green", alpha = 0.5, label = "record 2024")
plt.xlabel("Product Name")
plt.ylabel("record_2023 ammout in lakhs")
plt.title("Company Profit Info")
plt.legend()
```

```python
product_name = ["Hanky", "Shoose", "Creems", "Shirts", "Jeens"]
record_2023 = [20, 40, 30, 50, 40]
record_2024 = [30, 50, 40, 60, 80]
```

```python
plt.bar(product_name, record_2024)

for i in range(len(product_name)):
    plt.annotate(xy = (product_name[i], record_2024[i]), text = f"( {record_2024[i]} )")
```

```python
plt.bar(product_name, record_2023, color = "red", label = "record 2023")
plt.bar(product_name, record_2024, color = "green", alpha = 0.5, label = "record 2024", width = 0.3, align = "edge")
plt.xlabel("Product Name")
plt.ylabel("record_2023 ammout in lakhs")
plt.title("Company Profit Info")
plt.legend()

for i in range(len(record_2024)):
    plt.annotate(xy = (product_name[i], record_2023[i]), text = f"{record_2023[i]}")

for i in range(len(record_2024)):
    plt.annotate(xy = (product_name[i], record_2024[i]), text = f"{record_2024[i]}")
```

```python
calories = [30, 40, 30, 60, 20]
```
