```python
def genres_to_list(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i["name"])
    return L
```

```python
def cast_to_list(text):
    L = []
    x = 0
    for i in ast.literal_eval(text):
        if x < 3:
            L.append(i["name"])
            x = x + 1
        else:
            break
    return L
```

```python
def crew_to_list(text):
    L = []
    for i in ast.literal_eval(text):
        if i["job"] == "Director":
            L.append(i["name"])
    return L
```
