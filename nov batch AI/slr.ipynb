```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(cgpa, pkg, test_size = 0.2)
```

```python
# LinerRegression
```

```python
from sklearn.linear_model import LinearRegression
```

```python
mylr = LinearRegression()
```

```python
mylr.fit(x_train, y_train)
```

```python
# y = mx + b
```

```python
# m = slope
# b = intercept
```

```python
m = mylr.coef_
m
```

```python
b = mylr.intercept_
b
```

```python
x_test
```

```python
y_test
```

```python
y = m * 9.1 + b
y
```

```python
# prediction

cga = np.array(7)
cga_new = cga.reshape(-1, 1)

mylr.predict(cga_new)
```

```python
x_test
```

```python
y_pred = mylr.predict(x_test)
```

```python
y_pred
```

```python
# check accuracy
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_pred)
```

```python
plt.scatter(x_train, y_train, color = "red", label = "Training Data")
plt.scatter(x_test, y_test, color = "green", label = "Testing Data")
plt.plot(x_test, y_pred, label = "Predicted Line")
plt.legend()
plt.xlabel("CGPA")
plt.ylabel("Package")
plt.title("Placement Record")
```
